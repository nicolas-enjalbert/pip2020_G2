```python
url = "http://sabbar.fr/management/le-management-strategique-et-le-management-operationnel/#:~:text=Le%20management%20op%C3%A9rationnel%20correspond%20aux,pour%20atteindre%20les%20objectifs%20fix%C3%A9s."
```

```python
def getSoup(url):
    
    response = requests.get(url)
    print("statut de la reponse ", response.status_code)
    print('\n')
    soup = bs(response.text, 'html.parser')
    print(soup)
    
    return soup
```

```python
soup = getSoup(url)
```

```python
def getContent(soup):
    
    #Récupération du contenu de la page web (les paragraphes, avec et sans les balises html)
    art_content_html = soup.find_all('div', class_="entry-content")
    
    art_content = " ".join([x.text for x in art_content_html])
    art_content = art_content.replace("\n", "").replace("\xa0", "")

    return art_content, art_content_html
```

```python
def getDate(soup):
    
    art_extract_datetime = soup.find('script', class_ = "yoast-schema-graph yoast-schema-graph--main").content
    #art_extract_datetime = art_extract_datetime.replace("</script>","")
    return art_extract_datetime
```

```python
def getLang(soup):
    
    art_lang = soup.find('meta', property = "og:locale").get('content')

    return art_lang
```

```python
def getTitle(soup):
    
    art_title = soup.find('meta', property = "og:title").get('content')
    
    return art_title
```

```python
def getUrl(soup):
    
    art_url = soup.find('link', rel = 'canonical').get('href')
    
    return art_url
```

```python
def getSourceName(soup):
    return src_name
```

```python
def getSourceType(soup):
    return src_type
```

```python
def getSourceUrl(soup):
    
    src_url = soup.find('a', rel = 'home').get('href')
   
    return src_url
```

```python
print(getSourceUrl(soup))
```

```python
def getImg(soup):
    #Pas sur le site
    return src_img
```

```python
def getArtAuth(soup):
    #Pas sur le site
    return art_auth
```

```python
def getTag(soup):
    #Pas sur le site
    return art_tag
```

```python
def scrapPage(soup):
    
    #Récupération du contenu de la page web (les paragraphes, avec et sans les balises html)
    art_content, art_content_html = getContent(soup)
    
    #Extraction de la date de l'article
    art_extract_datetime = date.today().strftime("%d/%m/%Y")
    
    #Langue de l'article
    art_lang = getLang(soup)
    
    #Titre
    art_title = getTitle(soup)
    
    #Url
    art_url = getUrl(soup)
    
    #Nom de la source
    src_name = np.nan
    
    #Type de la source
    src_type = "xpath_source"
    
    #url source
    src_url = getSourceUrl(soup)
    
    #Image(s)
    src_img = np.nan 
    
    #Auteur de l'article
    art_auth = np.nan
    
    #Tag de l'auteur
    art_tag = np.nan
    
    return art_content, art_content_html, art_extract_datetime, art_lang, art_title, art_url, src_name, src_type, src_url, src_img, art_auth, art_tag
```
